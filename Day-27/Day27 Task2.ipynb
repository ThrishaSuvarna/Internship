```python
# Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression

# ---------------------------
# 1. Load Dataset
# ---------------------------
df = pd.read_csv("C:/Users/redmi/Downloads/hiring.csv")

print("Original Dataset:\n")
print(df)

# ---------------------------
# 2. Data Cleaning
# ---------------------------

# Convert experience words to numbers
experience_map = {
    "zero": 0, "one": 1, "two": 2, "three": 3,
    "four": 4, "five": 5, "six": 6, "seven": 7,
    "eight": 8, "nine": 9, "ten": 10,
    "eleven": 11, "twelve": 12
}

df['experience'] = df['experience'].fillna("zero")
df['experience'] = df['experience'].map(experience_map)

# Fill missing test score with median
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].median()
)

print("\nCleaned Dataset:\n")
print(df)

# ---------------------------
# 3. Prepare Data
# ---------------------------
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# ---------------------------
# 4. Train Model
# ---------------------------
model = LinearRegression()
model.fit(X, y)

# ---------------------------
# 5. Model Details
# ---------------------------
print("\nModel Coefficients:")
print("Experience Coef:", model.coef_[0])
print("Test Score Coef:", model.coef_[1])
print("Interview Score Coef:", model.coef_[2])
print("Intercept:", model.intercept_)

# ---------------------------
# 6. Predictions (No Warning Version)
# ---------------------------

# Candidate 1
candidate1 = pd.DataFrame([[2, 9, 6]],
                          columns=X.columns)

salary1 = model.predict(candidate1)
print("\nPredicted Salary (2 yr, 9 test, 6 interview):", salary1[0])

# Candidate 2
candidate2 = pd.DataFrame([[12, 10, 10]],
                          columns=X.columns)

salary2 = model.predict(candidate2)
print("Predicted Salary (12 yr, 10 test, 10 interview):", salary2[0])
```
